【問題1】1次関数
xが-50から50の範囲で次の式を考えます。

$$
y=\frac{1}{2}x+1
$$

-50から50まで0.1間隔のxを表すndarrayと、各要素に対するyを表すndarrayを作成してください。


```python
import numpy as np
np.set_printoptions(formatter={'float': '{:.1f}'.format})

x_ndarray = np.arange(-50,50.01,0.1)
print(x_ndarray)
print(x_ndarray.shape)

y_ndarray = (1/2)*x_ndarray+1
print(y_ndarray)
print(y_ndarray.shape)
```

```python
"""
【問題2】配列の結合
問題1で作成したndarrayを結合し、(1001, 2)の形の配列にしてください。0列目にx、1列目にyが来るようにします。
"""
xy_ndarray = np.vstack([x_ndarray,y_ndarray]) 
xy_ndarray = np.transpose(xy_ndarray)
print(xy_ndarray)
print(xy_ndarray.shape)
```

```python
"""
【問題3】勾配を求める
各x座標における勾配を求めます。ここで勾配はyの変化量をxの変化量で割ることで計算します。
"""
x_variation = xy_ndarray[1:,0]- xy_ndarray[:-1,0]


y_variation = xy_ndarray[1:,1]- xy_ndarray[:-1,1]
print(y_variation)
print("")

xy_gradient = y_variation / x_variation
print(xy_gradient)
```

```python
"""
【問題4】グラフを描く
1次関数のグラフをMatplotlibを用いて描いてください。
また、それとは別に横軸をx、縦軸を勾配としたグラフも描いてください
"""

import matplotlib.pyplot as plt

plt.plot(x_ndarray,y_ndarray)
plt.show()

plt.plot(x_ndarray[1:], xy_gradient)
plt.show()
```

```python
"""
【問題5】Pythonの関数化
問題1から問題3までの範囲をまとめて行うPythonの関数を雛形を参考に作成してください。
そして、勾配を0.1間隔で求めてndarrayで出力してください。
"""

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    変化量を使い勾配を計算する。
    Parameters
    ----------------
    function : function
      求めたい関数、xのndarrayを入れるとyのndarrayを返すもの。
    x_range : tuple
      np.arange()の引数と同じ方法で範囲を指定する。
      
    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      xとyを結合したもの。nはx_rangeに依存する。
    gradient : ndarray, shape(n-1,)
      関数の勾配。差をとるため要素は1つ減る
    """
    x_ndarray = np.arange(x_range[0],x_range[1],x_range[2])
    y_ndarray = function(x_ndarray)

    array_xy = np.vstack([x_ndarray,y_ndarray]) 
    array_xy = np.transpose(array_xy)
    
    x_variation = array_xy[1:,0]- array_xy[:-1,0]
    y_variation = array_xy[1:,1]- array_xy[:-1,1]
        
    gradient = y_variation / x_variation
    
    return array_xy, gradient

def function1(array_x):
    array_y = array_x * array_x
    return array_y

array_xy1, gradient1 = compute_gradient(function1)

plt.plot(array_xy1[:,0],array_xy1[:,1])
plt.show()

plt.plot(array_xy1[:-1,0], gradient1)
plt.show()

def function2(array_x):    
    array_y = 2 * (array_x**2) + 2 ** array_x
    return array_y

array_xy2, gradient2 = compute_gradient(function2)

plt.plot(array_xy2[:,0],array_xy2[:,1])
plt.show()

plt.plot(array_xy2[:-1,0], gradient2)
plt.show()

def function3(array_x):    
    array_y = np.sin(array_x**(1/2))
    return array_y

array_xy3, gradient3 = compute_gradient(function3,(0, 50.1, 0.1))

plt.plot(array_xy3[:,0],array_xy3[:,1])
plt.show()

plt.plot(array_xy3[:-1,0], gradient3)
plt.show()
```

```python
"""
【問題6】最小値を求める
問題5で扱った関数のyの最小値を求めます。xの範囲も問題5と同様とします。
今回は勾配降下法ではなく、ndarrayのminメソッド、argminメソッドを用いてください。
minメソッドは最小値を、argminメソッドは最小値のインデックスを取得することができます。
また、最小のyをとる前後の勾配を表示してください。
"""
print("Function1関数のyの最小値は{}で、そのインデックスは{}で、その時の勾配は{}です。".format((min(array_xy1[:,1])),(np.argmin(array_xy1[:,1])),(gradient1[np.argmin(array_xy1[:,1])])))
print("Function2関数のyの最小値は{}で、そのインデックスは{}で、その時の勾配は{}です。".format((min(array_xy2[:,1])),(np.argmin(array_xy2[:,1])),(gradient2[np.argmin(array_xy2[:,1])])))
print("Function3関数のyの最小値は{}で、そのインデックスは{}で、その時の勾配は{}です。".format((min(array_xy3[:,1])),(np.argmin(array_xy3[:,1])),(gradient3[np.argmin(array_xy3[:,1])])))

```
