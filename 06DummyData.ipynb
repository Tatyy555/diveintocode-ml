```python
"""
【課題】ダミーデータの作成
【問題1】乱数の作成

"""

import numpy as np

mean = (-3, 0) #行の平均を意味している。ランダムで作成した行列の中心部分。
cov = [[1.0, 0.8], [0.8, 1.0]] #分散と共分散。分散は1に近付くほど正の相関があるとされ、共分散はそれが1に近付くほどバラつき度合が低いよう。
data = np.random.multivariate_normal(mean, cov, size=500)
np.random.seed(0)

print(mean)
print(cov)
print(data.shape)
print(data)


"""
【問題2】散布図による可視化
内容：matplotlibのplt.scatter()を使う。
"""

import matplotlib.pyplot as plt
plt.scatter(data[:,0],data[:,1])
plt.show()
```

```python
"""
【問題3】ヒストグラムによる可視化
ヒストグラムはplt.hist()を使うことで描け、
それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えることができる。
"""

plt.hist(data[:,0],bins=100)
plt.xlim(-6,3)
plt.show()
plt.hist(data[:,1],bins=100)
plt.xlim(-6,3)
plt.show()

```

```python
"""
【問題4】データの追加
問題1から平均が(0, -3)だけ違う乱数をもう500個作成し、
それぞれのデータをひとつの散布図として可視化する。
凡例として問題1のものは0、問題4のものは1とする。
"""

mean = (0, -3) 
cov = [[1.0, 0.8], [0.8, 1.0]] 
data2 = np.random.multivariate_normal(mean, cov, size=500)
np.random.seed(1)

plt.scatter(data[:,0],data[:,1], label="0")
plt.scatter(data2[:,0],data2[:,1], label="1")
plt.legend()
plt.show()
```

```python
"""
【問題5】データの結合
問題1、4で作成したndarrayを結合し、(1000, 2)のndarrayとしてください。
結合はnp.concatenate()やnp.vstack()を使うことで行えます。
"""

data3 = np.vstack((data, data2))
print(data3.shape)
print(data3)
```

```python
"""
【問題6】ラベル付け
ひとまとめになった1000個のデータそれぞれに対して、
問題1、4どちらで作成したものなのかを示すラベル付けを行ってください。
問題1のものには0、問題4のものには1を対応させます。
0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。
機械学習に使用するデータセットはこのような形になっていることが多いです。
"""

label0 = np.full((500,1),0)
print(label0.shape)

label1 = np.full((500,1),1)
print(label1.shape)

label01 = np.vstack((label0,label1))
print(label01.shape)

data4 = np.hstack((data3,label01))
print(data4.shape)
print(data4)
```
