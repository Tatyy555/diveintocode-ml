```python
"""
【問題1】100日目の米粒の数
100日目には合計で米粒をいくつもらっているのかを計算するコードを作成し、
答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と
「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）
"""
grain = 1
n_days = 10
total_grains = 0
grain_on_the_day = []
total_grains_on_the_day = []
for day in range(1,n_days+1):
        grains = grain*2**(day-1)
        total_grains = total_grains + grains
        grain_on_the_day.append(grains)
        total_grains_on_the_day.append(total_grains)
print("{}日目にもらう米の数： {}粒".format(n_days,grains))
print("{}日目までに累計で貰う米の数：{}粒".format(n_days, total_grains))

import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("Grains on the Day",{"fontsize":20})
plt.xlabel("Days",{"fontsize":20})
plt.ylabel("Grains",{"fontsize":20})
plt.plot(grain_on_the_day, color='green', lw=5)
plt.tick_params(labelsize=20)
plt.show()

plt.title("Total Grains on the Day",{"fontsize":20})
plt.xlabel("Days",{"fontsize":20})
plt.ylabel("Grains",{"fontsize":20})
plt.plot(total_grains_on_the_day, color='blue', lw=5)
plt.tick_params(labelsize=20)
plt.show()
```

```python
"""
【問題3】貰える米で何人が何日生活できるか
この米があれば何人が何日生活できるか計算する関数を作成してください。
コースを受講する期間に貰える米粒で、このクラスにいる人数であれば、何日間生活できるかを出力してみましょう。
関数の最低限の入力は 米粒の数(int)と人数(int)、出力は生活できる日数(int)です。
関数名やコメントなどは分かりやすく考えて作成してください。

方針：コースを受講する期間を3カ月＝90日とする。
      問題②よりコースを受講する期間(90日)で貰える米粒の数は1237940039285380274899124223粒だった。
　　　クラスの人数は5名（生徒数）とする。
   　 一人当たりの一日の米粒量は13,000粒とする
        ・茶碗中盛 3,250粒(150g)（農林水産省WEB　https://www.maff.go.jp/j/heya/sodan/1806/03.html）
        ・摂取目安量としては1日に茶碗3杯もしくは4杯が良いとされている（くりや株式会社　https://kuriya.co.jp/blog/other/20180520-576.html）
        ・したがって、3,250粒×4杯(多めに計算) = 13,000粒とする。
"""


def how_many_days_survive(n_grains=1237940039285380274899124223,n_members=5):
    """
    コースを受講する期間に貰える米粒で、このクラスにいる人数であれば、何日間生活できるか計算する関数。
    
    Parameteres
    -----------
    n_grains: int
        米粒の数（default: 1237940039285380274899124223）
    n_members: int
        クラスにいる生徒数（default: 5）
    
    Returns
    -------
    n_days_survive: int
        生活できる日数
        
    """
    n_days_survive = n_grains / (n_members * 13000 * 3) 
    print("コースを受講する期間(90日)に貰える米粒で、このクラスにいる{}人であれば、{}日間生活できる。".format(n_members, n_days_survive))
    return n_days_survive


n_days_survive = how_many_days_survive()
```
