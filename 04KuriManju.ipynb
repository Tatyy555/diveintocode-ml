```python
"""
課題：栗まんじゅう問題
問題：1個の栗まんじゅうにバイバインを振りかけた場合、
　　　太陽系が栗まんじゅうで覆われるのは何分後でしょうか。これを計算するプログラムを作成してください。
　　　栗まんじゅうや太陽系の体積を調査した上で仮定して計算してください。
   　 栗まんじゅうの隙間や重力などは考えず、問題を単純化して構いません。
      ボールにバイバインを振りかけた時、東京ドームが一杯になる時間など他の状況にも
      適用できるような汎用的なプログラムを作成してください。
      また、グラフを表示するなど分かりやすいノートブックを作成してください。

方針：栗まんじゅうは5分後に体積ば倍になると考える

      栗まんじゅうの体積 = 縦 * 横 * 高さ
                         = 0.05m * 0.05m * 0.04m
                         = 100cm^3
                         = 0.0001m^3
　　　
      太陽系の体積は、太陽系の中心である太陽から最も遠い海王星を半径とした球体と考える
   　 
      太陽⇒海王星の距離 = 30.01au
     　　　　　　　　　　= 30.01 * 149597870700m 
                         = 4489432099707m
     
   　 太陽系の体積     　= 4 * π * 半径^3 / 3
                         = 4 * 3.14159265359 * (4489432099707m)^3 / 3
                         = 3.790206147800111e+38
   
   
   　 東京ドームの体積   = 1240000m^3

参考：「wiki 太陽系」https://ja.wikipedia.org/wiki/%E5%A4%AA%E9%99%BD%E7%B3%BB
　　　「wiki 天文単位」https://ja.wikipedia.org/wiki/%E5%A4%A9%E6%96%87%E5%8D%98%E4%BD%8D
   　 「wiki 東京ドーム」https://ja.wikipedia.org/wiki/%E6%9D%B1%E4%BA%AC%E3%83%89%E3%83%BC%E3%83%A0_(%E5%8D%98%E4%BD%8D)
   

計算式：求めたい体積 = 栗まんじゅうの体積 ** 経過した時間(分)/5分

"""

def kuri_manju_bibin(kuri_manju_volume = 0.0001, goal_volume = 3.790206147800111e+38):
    """
    Parameteres
    -----------
    kuri_manju_volume: float
        栗まんじゅうの体積（default: 0.0001㎥）
    goal_vlolume: int
        求めたい体積（default: 3.790206147800111e+3㎥）
    
    Returns
    -------
    elapsed_time: int
        経過した時間（分）
    """
    
    elapsed_time = 0
    list_kuri_manju_volume = [kuri_manju_volume]
    list_elapsed_time = [elapsed_time]
    while(goal_volume >= kuri_manju_volume):
        kuri_manju_volume = kuri_manju_volume * 2
        elapsed_time = elapsed_time + 5
        print(str(elapsed_time)+"分経ちました。")
        print("栗まんじゅうの総体積は"+str(kuri_manju_volume)+"㎥になりました。")
        list_kuri_manju_volume.append(kuri_manju_volume)
        list_elapsed_time.append(elapsed_time)
    
    print(str(elapsed_time)+"分経過して栗まんじゅうの体積が"+str(kuri_manju_volume)+"㎥となり、")
    print("想定していた容量の"+str(goal_volume)+"㎥を超えるのに{}分掛かりました。".format(elapsed_time))
    
    import matplotlib.pyplot as plt
    %matplotlib inline
    plt.title("Volume of Kuri Manju",{"fontsize":20})
    plt.xlabel("Minutes",{"fontsize":20})
    plt.ylabel("Volume",{"fontsize":20})
    plt.plot(list_elapsed_time,list_kuri_manju_volume,  color='blue', lw=5)
    plt.tick_params(labelsize=20)
    plt.show()

    return elapsed_time

print("太陽系が栗まんじゅうで一杯になるまでの時間")
elapsed_time = kuri_manju_bibin()
print()

print("東京ドームが栗まんじゅうで一杯になるまでの時間")
elapsed_time = kuri_manju_bibin(goal_volume = 1240000)
```
