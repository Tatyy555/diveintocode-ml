```python
"""
課題：1枚の紙を何回折り曲げたら目的の厚さになるか計算するプログラムの作成
数式：t = 折り曲げた時の厚さ
　　　n = 折り曲げた回数
　　　tn = t0 × 2^n
　　　t0 = 0.00008m

【問題1】何回折ったら富士山を超えるか
内容：厚さが「富士山の高さ(3776m)」を超えるのに必要な紙を折る最小回数を計算するプログラムを作成
"""

thickness = 0.00008
goal_thickness = 3776
n_fold = 0
while(goal_thickness >= thickness):
    thickness = thickness * 2
    n_fold = n_fold + 1
    print(str(n_fold)+"回折りました。")
    print(str(thickness)+"mに到達しました。")

print("紙を"+str(n_fold)+"回折って厚さが"+str(thickness)+"mとなり、目的の厚さの"+str(goal_thickness)+"mを超えたので折るのを止めます。")
print("目的の厚さを超えるために紙を折った回数は{}回でした".format(n_fold))
```

```python
"""
【問題2】任意の厚さに対応した関数
内容：高さを入力すると、それを超えるのに必要な紙を折る最小回数を出力する関数の作成。
　　　「最も近い太陽以外の恒星」に到達するまでに紙を折る回数の計算。
数式：t0 = 任意
　　　最も近い太陽以外の恒星 = プロキシマ・ケンタウリ
                             = 4.0175 * 10^16 
"""

def how_many_times_to_fold(thickness = 0.00008, goal_distance = 3776):
    n_fold = 0
    while(goal_distance >= thickness):
        thickness = thickness * 2
        n_fold = n_fold + 1
        print(str(n_fold)+"回折りました。")
        print(str(thickness)+"mに到達しました。")
    print("紙を"+str(n_fold)+"回折って厚さが"+str(thickness)+"mとなり、目的の厚さの"+str(goal_distance)+"mを超えたので折るのを止めます。")
    print("目的の厚さを超えるために紙を折った回数は{}回でした".format(n_fold))
    return n_fold, thickness

n_fold, thickness = how_many_times_to_fold(goal_distance = 4.0175 * 10**16)

        
```
